### Boxplots

Perform a check for outliers on relevant columns such as trip distance and trip duration. Remember, some of the best ways to identify the presence of outliers in data are box plots and histograms. 

**Note:** Remember to convert your date columns to datetime in order to derive total trip duration.  

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

**trip distance**

```python
plt.figure(figsize=(8,2))
sns.boxplot(df, x='trip_distance', fliersize=3).set_yticks([])
plt.xlabel('Miles')
plt.title('Trip Distance')
plt.plot()
```

```python
plt.figure(figsize=(7,4.2))
sns.histplot(df, x='trip_distance', bins=range(0,26,1))
plt.ylabel('Count')
plt.xlabel('Trip Distance')
plt.title('Trip distance histogram')
plt.show()
```

**total amount**

```python
plt.figure(figsize=(8,3))
sns.boxplot(df, x='total_amount', fliersize=3).set_yticks([])
plt.xlabel('USD')
plt.title('Total_Amount')
plt.show()
```

```python
plt.figure(figsize=(8,4.2))
sns.histplot(df, x='total_amount', bins=range(0,26,1))
plt.xlabel('USD')
plt.ylabel('Count')
plt.title('Total Amount')
plt.show()
```

**tip amount**

```python
plt.figure(figsize=(8,3))
sns.boxplot(df, x='tip_amount', fliersize=3).set_yticks([])
plt.xlabel('USD')
plt.title('Tip Amount')
plt.show()
```

```python
plt.figure(figsize=(12, 4.5))
sns.histplot(df, x='tip_amount', bins=range(0,21,1)).set_xticks(range(0,21,2))
plt.xlabel('USD')
plt.ylabel('Count')
plt.title('Tip Amount')
plt.show()
```

**tip_amount by vendor**

```python
plt.figure(figsize=(12, 4.5))
sns.histplot(df, x='tip_amount', hue='VendorID', palette='Blues', multiple='stack', bins=range(0,21,1)).set_xticks(range(0,21,2))
sns.set_theme(style='darkgrid')
plt.xlabel('USD')
plt.ylabel('Count')
plt.title('Tip Amount')
plt.show()
```

Next, zoom in on the upper end of the range of tips to check whether vendor one gets noticeably more of the most generous tips.

```python
# Create histogram of tip_amount by vendor for tips > $10 
#==> ENTER YOUR CODE HERE

high_tips = df[df['tip_amount'] > 10]
plt.figure(figsize=(12, 4.5))
sns.histplot(high_tips, x='tip_amount', hue='VendorID', palette='Blues', multiple='stack', bins=range(10,21,1)).set_xticks(range(10,21,2))
plt.xlabel('USD')
plt.ylabel('Count')
plt.title('Tip Amount')
plt.show()
```

**Mean tips by passenger count**

Examine the unique values in the `passenger_count` column.

```python
df.groupby('passenger_count').size().sort_values(ascending=False)
```

```python
tipbar = df.groupby('passenger_count')['tip_amount'].mean().reset_index()
tipbar = tipbar.set_index('passenger_count')
tipbar
```

```python
# Create bar plot for mean tips by passenger count
#==> ENTER YOUR CODE HERE
data = tipbar.tail(-1)
plt.figure(figsize=(11,6))
sns.barplot(data, x='passenger_count', y='tip_amount').set_yticks(range(0,3,1))
plt.xlabel('Passenger Count')
plt.ylabel('Tip Amount')
plt.title('Average Tip Amount Per Passenger Count')
for n, v in enumerate(data['tip_amount']):
    plt.text(n, v + 0.02, f'{v:.2f}', ha='center')
plt.axhline(df['tip_amount'].mean(), ls='--', color='red', label='global mean')
plt.legend()
plt.show()
```

**Create month and day columns**

```python
df['month'] = df['tpep_pickup_datetime'].dt.month_name()

df['day'] = df['tpep_pickup_datetime'].dt.day_name()
```

**Plot total ride count by month**

Begin by calculating total ride count by month.

```python
# Get total number of rides for each month
#==> ENTER YOUR CODE HERE
monthly_rides = df.groupby('month').size()
monthly_rides
```

Reorder the results to put the months in calendar order.

```python
# Reorder the monthly ride list so months go in order
#==> ENTER YOUR CODE HERE
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

monthly_rides = monthly_rides.reindex(index=month_order)
monthly_rides
```

```python
# Show the index
#==> ENTER YOUR CODE HERE
monthly_rides.index
```

```python
# Create a bar plot of total rides per month
#==> ENTER YOUR CODE HERE
plt.figure(figsize=(13,6))
sns.barplot(x=monthly_rides.index, y=monthly_rides, hue=monthly_rides.index)
plt.xlabel('')
plt.ylabel('')
plt.title('Ride count by month', fontsize=16)

for n,v in enumerate(monthly_rides):
    plt.text(n, v + 20, v, ha='center')

plt.show()
```

**Plot total ride count by day**

Repeat the above process, but now calculate the total rides by day of the week.

```python
rides_day = df.groupby('day').size()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

rides_day = rides_day.reindex(index=day_order)

rides_day
```

```python
plt.figure(figsize=(11,5))
sns.barplot(x=rides_day.index, y=rides_day, hue=rides_day, legend=False)
plt.xlabel('')
plt.ylabel('')
plt.title('Ride count by day', fontsize=16)

for n,v in enumerate(rides_day):
    plt.text(n, v + 30, v, ha='center')


plt.show()
```

**Plot total revenue by day of the week**

Repeat the above process, but now calculate the total revenue by day of the week.

```python
revenue_day = df.groupby('day')['total_amount'].sum()
revenue_day = revenue_day.reindex(index=day_order)
revenue_day
```

```python
# Create bar plot of total revenue by day
#==> ENTER YOUR CODE HERE

plt.figure(figsize=(11,6))
sns.barplot(x=revenue_day.index, y=revenue_day, hue=revenue_day.index)
plt.title('Total Revenue by Day')
plt.xlabel('')
plt.ylabel('USD')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K'))
for n, v in enumerate(revenue_day):
    plt.text(n, v + 500, f'{v:.0f}', ha='center')
plt.show()
```

**Plot total revenue by month**

```python
# Repeat the process, this time for total revenue by month
#==> ENTER YOUR CODE HERE

revenue_month = df.groupby('month')['total_amount'].sum()
revenue_month = revenue_month.reindex(index=month_order)
revenue_month
```

```python
# Create a bar plot of total revenue by month
#==> ENTER YOUR CODE HERE

plt.figure(figsize=(11,6))
sns.barplot(x=revenue_month.index, y=revenue_month, hue=revenue_month.index)
plt.title('Total Revenue by Month')
plt.xlabel('')
plt.ylabel('')

for n, v in enumerate(revenue_month):
    plt.text(n, v + 400, f'{v:.0f}', ha='center')

plt.show()
```

**Histogram of rides by drop-off location**

First, check to whether the drop-off locations IDs are consecutively numbered. For instance, does it go 1, 2, 3, 4..., or are some numbers missing (e.g., 1, 3, 4...). If numbers aren't all consecutive, the histogram will look like some locations have very few or no rides when in reality there's no bar because there's no location. 

```python
# Check if all drop-off locations are consecutively numbered
#==> ENTER YOUR CODE HERE
df['DOLocationID'].max() - len(set(df['DOLocationID']))
```

To eliminate the spaces in the historgram that these missing numbers would create, sort the unique drop-off location values, then convert them to strings. This will make the histplot function display all bars directly next to each other. 

```python
plt.figure(figsize=(13,5))
sorted_DOID = df['DOLocationID'].sort_values()
sorted_DOID = sorted_DOID.astype(str)
# Convert to string

sns.histplot(sorted_DOID, bins=range(0, df['DOLocationID'].max()+1, 1))
plt.xticks([])
plt.xlabel('Drop Off Location')
plt.title('Histogram of rides by drop-off location')
plt.show()
```
